```python
def process_object_columns(dataframe):
    object_columns = [col for col in dataframe.columns if dataframe[col].dtype =='object']
    
    def extract_second_part(text):
        if isinstance(text,str):
            parts = text.split('/')
            if len(parts) > 1:
                return parts[1]
            return text
    
    for column in object_columns:
        dataframe[column] = dataframe[column].apply(extract_second_part)
    return dataframe
```

```python
def process_object_columns(dataframe, columns_to_exclude=None):
    if columns_to_exclude is None:
        columns_to_exclude = []  
    
    object_columns = [col for col in dataframe.columns if dataframe[col].dtype == 'object']
    
    def extract_second_part(text):
        if isinstance(text, str):
            parts = text.split('/')
            if len(parts) > 1:
                return parts[1]
            return text
    
    for column in object_columns:
        if column not in columns_to_exclude:
            dataframe[column] = dataframe[column].apply(extract_second_part)
            
    return dataframe
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
# H0: There is no association between gender and the frequency of feeling depressed, stressed, or anxious. 
# H1: There is an association between gender and the frequency of feeling depressed, stressed, or anxious.

# Create a contingency table
contingency_table = pd.crosstab(updated_dataframe["What is your gender? "], updated_dataframe["Do you often feel depressed, stressed, or anxious? "])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the chi-square test result
print("Chi-Square Value:", chi2)
print("P-value:", p)
```

```python
# H0: There is no linear correlation between the number of hours of sleep and the number of days feeling anxious.
# H1: There is a linear correlation between the number of hours of sleep and the number of days feeling anxious.

# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(updated_dataframe['How many hours do you sleep on average per day? '], updated_dataframe['How many days during the last week did you feel: sad, depressed, stressed, or anxious? '])

# Print the results
print("Pearson Correlation Coefficient:", correlation_coefficient)
print("P-value:", p_value)
```

P-value: The p-value of approximately 0.0012 is less than the commonly used significance level of 0.05. This indicates that the observed correlation is statistically significant. In other words, the likelihood of observing this correlation by random chance alone is very low.

There is a statistically significant weak negative linear correlation between the number of hours of sleep and the number of days feeling anxious.
The negative correlation coefficient suggests that, on average, individuals who report more hours of sleep tend to report fewer days feeling anxious, and vice versa.
